# MatPlotlib
MatPlotlib is plotting library and numerical mathematics library extension to numpy.


```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#plotting using scatter

plt.scatter(x,y,c='black')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,x*x,'r--')
```

```python
#creating subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*-')
```

```python
#sine curve

x = np.arange(0,4* np.pi,0.1)
y = np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
```

```python
# subplots with sine and cos curve
x = np.arange(0,4* np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.title('sine wave form')
plt.plot(x,y_sin)
plt.subplot(2,2,2)
plt.title('cos wave form')
plt.plot(x,y_cos)

plt.show()
```

```python
# bar plot 

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('bar graph')

plt.show()
```

```python
#Box plot 

data = [np.random.normal(0,std,100) for std in range (1,4)]
plt.boxplot(data,vert=True, patch_artist=True)

plt.show()
```

```python
#Pie chart

labels = 'Pune','Banglore','Mumbai','Chennai'
size = [200,300,400,100]
colors = ['red','blue','yellow','green']
explode=(0,0,0,0)

plt.pie(size,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
