# Seaborn
1. distplot
2. jointplot
3. pairplot

```python
import seaborn as sb
```

```python
sns.set_theme()
```

```python
# tip is dependent on other factors
# tip is dependent feature
# other attributes are independent feature

df = sb.load_dataset('tips')
df.head()
```

# Correlation between Heatmap
It uses colored cells to show 2D correlation matrix between two discrete dimensions or event types. It is 
very important in Feature Selection


```python
df.corr()
```

```python
sb.heatmap(df.corr())
```

# JointPlot
Allows to study relationship between 2 numeric values. Generally used for Univarient analysis.
It is usually scatterplot, hexbin plot, 2D histogram or 2D density plot


```python
sb.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

# Pairplot
Known as scatterplot in which one variable in the same data row is matched with another 

```python
sb.pairplot(df)
```

```python
sb.pairplot(df,hue='sex')
```

# Distplot
It helps us to check the distribution of column feature

```python
sb.distplot(df['tip'])
```

```python
# kde -> kernal density estimation

sb.distplot(df['tip'],kde=False,bins=10)
```

# Categorial Plots
1. boxplot
2. violinplot
3. countplot
4. bar plot


```python
#countplot 

sb.countplot('day',data=df)
```

```python
#countplot 

sb.countplot(y='sex',data=df)
```

```python
# barplot

sb.barplot('total_bill',y='sex',data=df)
```

```python
#box plot

sb.boxplot('day','total_bill',data=df)
```

```python
# categroize data on other categories

sb.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
# violin plot
# It helps us to see distribution in kernel density and box plot

sb.violinplot(x='total_bill',y='day',data=df)
```

**Relationship plot** 

```python
sns.relplot(data=df,x='total_bill',y='tip',col='time',hue='smoker',style='smoker', size='size')
```
