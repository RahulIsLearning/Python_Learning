```python
import pandas as pd
```

```python
tata_company = pd.DataFrame({
    "variant":["nexon","altroz","tiago","tigor","safari"],
    "model":["base","classic","premium","classic","premium"],
    "body_type":["SUV","Hatchback","normal","sedan","SUV"]
})
tata_company
```

```python
hyundai_company = pd.DataFrame({
    "variant":["venue","i20","i10","verna","creata"],
    "model":["base","classic","premium","classic","premium"],
    "body_type":["SUV","Hatchback","normal","sedan","SUV"]
})
hyundai_company
```

```python
concat_df = pd.concat([tata_company,hyundai_company], ignore_index=True)
concat_df
```

```python
concat_df1 = pd.concat([tata_company,hyundai_company], keys=['tata','hyundai'])
concat_df1
```

```python
concat_df1.loc["tata"]
```

**Concat the two dataframes having different columns**

```python
windspeed_df = pd.DataFrame({
    "city":["pune","mumbai"],
    "windspeed":[10,34]
}, index=[1,0])
windspeed_df
```

```python
tempt_df = pd.DataFrame({
    "city":["mumbai","pune","solapur"],
    "tempreature":[28,26,30]
}, index=[0,1,2])
tempt_df
```

```python
concat_Df = pd.concat([tempt_df, windspeed_df], axis = 1)
concat_Df
```
