In this we will use below 3 methods to handle missing values in given data

1. **fillna()** to fill default having missing values
2. interpolate to make guess on missing values
3. **dropna()** to drop the rows with missing values

```python
# raw_data
input_data = pd.read_csv('titanic_train.csv')
input_data.set_index('PassengerId', inplace=True)  # setting the index to passengerId
rows,columns = input_data.shape
print('rows',rows)
print('columns',columns)
```

```python
# raw sample_data
sample_data = input_data.head(10)
sample_data
```

**Handling missing String with fillna() method**

```python
# fillna() method

sample_data.fillna(0) # This will fill all the null/missing values with 0. It is not suitable when we have different datatypes 
# as missing values
```

```python
# fillna() method for different datatypes

new_df = sample_data.fillna({
    'Age':0.0,
    'Cabin': 'No Cabin'
})
new_df
```

```python
# ffill() method for filling missing values

# It will carryforward the previous data as filler to missing data

forwardfill = sample_data.fillna(method='ffill')
forwardfill
```

```python
# bfill() method for filling missing values

# It will carrybackword the next data as filler to missing data

backwardfill = sample_data.fillna(method='bfill')
backwardfill
```

**Handling missing values for interpolate method**

```python
interpolate_data = sample_data.interpolate()
interpolate_data
```

**Dropping the records having the missing values**

```python
drop_data = sample_data.dropna()
drop_data
```
